# BIBLIOTECAS

```python
import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# LEITURA DOS DADOS

```python
url = 'https://raw.githubusercontent.com/RyanXHF/ml_finished_projects/main/Customer-Churn.csv'  # Use raw content URL
db = pd.read_csv(url)
db
```

## RENOMEANDO COLUNAS

```python
troca = {'Nao': 0,
         'Sim': 1} # converte sim/nao em 1/0

db['Churn'] = db.Churn.map(troca) # realiza a conversão sim/nao em 1/0
db
```

```python
x = db[['Maior65Anos', 'Conjuge', 'Dependentes', 'MesesDeContrato', "TelefoneFixo",
        "VariasLinhasTelefonicas",	"ServicoDeInternet",	"SegurancaOnline",	"BackupOnline",
        "SeguroNoDispositivo",	"SuporteTecnico",	"TVaCabo",	"StreamingDeFilmes",
        "TipoDeContrato", "PagamentoOnline",	"FormaDePagamento",	"ContaMensal"]]	# Features

x
```

```python
y = db["Churn"] # Classes
y
```

# SELEÇÃO DOS DADOS PARA TREINO

CLASSIFICAÇÃO BINARIA
* CLASSES (Y) -> bought(1/0) -> CLASSIFICAÇÃO BINARIA
* FEATURE NÃO DEFINE SE CLASSE É BINARIA


- CONSTRUÇÃO DO MODELO MACHINE LEARNING
   * TREINO 75% DADOS -> FIT
   * TESTE 25% -> ACURACIA

```python
# Separar treino e teste
seed = 20
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, random_state=seed, stratify=y)

print('Treinamento com %d elementos e testarmos com %d elementos' % (len(train_x), len(test_x)))
```

### MODELO CLASSIFICADOR

```python
# Import necessary library
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Iterate through columns of train_x and test_x and encode string columns
for col in train_x.columns:
    if train_x[col].dtype == 'object':
        train_x[col] = label_encoder.fit_transform(train_x[col])
        test_x[col] = label_encoder.transform(test_x[col]) # Use the same encoder to transform test_x


model = LinearSVC() # Cria o modelo
model.fit(train_x, train_y) # treino

result = model.predict(test_x) # teste

acuracia = accuracy_score(test_y, result) # compara vetores
acuracia_model = model.score(test_x, test_y) # roda teste

print("Acuracia: ", (acuracia*100), "%")
print("Acuracia_model: ", (acuracia_model*100), "%")
```

### DUMMY (BASELINE)

```python
from sklearn.dummy import DummyClassifier

dummy_stratified = DummyClassifier()
dummy_stratified.fit(train_x, train_y)
acuracia_dummy = dummy_stratified.score(test_x, test_y)

print("Acuracia Dummy: ", (acuracia_dummy*100), "%")
```
